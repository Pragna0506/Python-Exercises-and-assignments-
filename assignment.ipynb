- For and to be True, both conditions must be True. If the first one is already False, checking the second is unnecessary

2. How does the or operator behave if the first operand evaluates to True?
- Answer:
If the first operand is True, Python does not evaluate the second operand and immediately returns True
- Why?
For or, only one condition needs to be True.

```python
#1. Elements in Either Set but Not Both
set_a = {1, 2, 3}
set_b = {3, 4, 5}

result = set_a.symmetric_difference(set_b)

print(result)
#Explanation:
#symmetric_difference() returns elements that are in either set but not in both.
```

```python
#2. Difference Between .copy() and =
#Using .copy()
original = {1, 2, 3}
copy_set = original.copy()

original.add(4)

print(original)
print(copy_set)

# Using Assignment =
original = {1, 2, 3}
assigned_set = original

original.add(4)

print(original)
print(assigned_set)

#Explanation:
#.copy() creates a new set.
# = creates another reference to the same set.

```
